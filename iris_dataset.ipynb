```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
dir(df)
```

```python
colums = df.feature_names
colums
```

```python
df.data[0]
```

```python
df.target
```

```python

_, ax = plt.subplots()
scatter = ax.scatter(df.data[:, 0], df.data[:, 1], c=df.target)
ax.set(xlabel=df.feature_names[0], ylabel=df.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], df.target_names, loc="lower right", title="Classes"
)
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(df.data[:, 2], df.data[:, 3], c=df.target)
ax.set(xlabel=df.feature_names[2], ylabel=df.feature_names[3])
_ = ax.legend(
    scatter.legend_elements()[0], df.target_names, loc="lower right", title="Classes"
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
xtrain,xtest,ytrain,ytest = train_test_split(df.data, df.target, test_size = 0.2 , random_state = 42)
model = LogisticRegression(max_iter = 1000)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
model.predict([[5.5, 3.0 , 4.4, 1.3]])
```

```python
model.predict([[2.5, 6.5 , 3.4, 7.3]])
```

```python
y_predict = model.predict(xtest)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest , y_predict)
cm
```

```python
import seaborn as sns
plt.figure(figsize= (6,4))
sns.heatmap(cm , annot = True)
plt.xlabel("predicted")
plt.ylabel("truth")
```
